# (Paso 4) ⚙️ Preparando las imágenes: el arte de limpiar datos

Las imágenes del dataset vienen en distintos tamaños, colores y formatos. Nuestra red neuronal necesita recibir **siempre el mismo tipo de entrada**. Es como preparar ingredientes antes de cocinar: todo debe estar cortado del mismo tamaño.

**¿Qué haremos en esta celda?**

1. **Redimensionar** → Todas las imágenes al mismo tamaño: 100×100 píxeles
2. **Escala de grises** → Convertir de color (RGB, 3 canales) a gris (1 canal). Esto reduce la cantidad de datos a procesar
3. **Normalizar** → Dividir entre 255. Los píxeles pasan de valores 0-255 a valores 0-1. Esto hace que la red neuronal aprenda más rápido
4. **Organizar en X e Y** → `X` = las imágenes (lo que la IA ve), `Y` = las etiquetas (la respuesta correcta)

```
Imagen original (300×400, color) → Preprocesada (100×100, gris, normalizada)
Valor de píxel: 200              →  200 / 255 = 0.784
```

> ⚠️ **Esta celda tarda varios minutos** porque procesa más de 23,000 imágenes. ¡Es normal! En la industria, el preprocesamiento puede tardar días para datasets masivos.

````python
# 📊🔍 Preparamos y visualizamos los datos: mostramos 25 imágenes del dataset en escala de grises y redimensionadas,
# luego procesamos todas las imágenes, las normalizamos y las organizamos en X (características) e Y (etiquetas) para el entrenamiento. 🚀🐶🐱
plt.figure(figsize=(20,20))
tamano_img = 100
datos_entrenamiento = []
X = []
Y = []
for i, (imagen, etiqueta) in enumerate(datos['train'].take(25)):
    imagen = cv2.resize(imagen.numpy(), (tamano_img, tamano_img))
    imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    plt.subplot(5,5,i+1)
    plt.imshow(imagen, cmap='grey')
    #plt.title('Perro' if etiqueta == 1 else 'Gato')
    plt.axis('off')
for i, (imagen, etiqueta) in enumerate(datos['train']):
  imagen = cv2.resize(imagen.numpy(), (tamano_img, tamano_img))
  imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
  imagen = imagen.reshape(tamano_img, tamano_img, 1)
  datos_entrenamiento.append([imagen, etiqueta])
for imagen, etiqueta in datos_entrenamiento:
  X.append(imagen)
  Y.append(etiqueta)
X = np.array(X).astype(float) / 255
Y = np.array(Y)

print(f'\n✅ Dataset listo para entrenar')
print(f'   Forma de X (imágenes): {X.shape}  → {X.shape[0]} imágenes de {X.shape[1]}x{X.shape[2]} px')
print(f'   Forma de Y (etiquetas): {Y.shape}')
print(f'   Gatos (0): {(Y==0).sum():,}  |  Perros (1): {(Y==1).sum():,}')
````

# (Paso 5) 🧠 Construyendo el cerebro de nuestra IA: Redes Neuronales Convolucionales

Aquí viene la parte más emocionante: **diseñar la arquitectura de nuestra red neuronal**.

## ¿Cómo ve una imagen una computadora?

Para una computadora, una imagen de 100×100 es simplemente una matriz de **10,000 números**. Cada número representa el brillo de ese píxel (0 = negro, 255 = blanco).

## ¿Cómo aprende a "ver" una CNN?

Una **Red Neuronal Convolucional (CNN)** usa filtros que van recorriendo la imagen buscando patrones:

```
Capa 1 (Conv2D 32 filtros)  → Detecta bordes y líneas básicas
Capa 2 (Conv2D 64 filtros)  → Detecta formas: círculos, esquinas
Capa 3 (Conv2D 128 filtros) → Detecta partes: orejas, hocico, ojos
Capa Dense (250 neuronas)   → Combina todo y decide: ¿Perro o Gato?
```

**¿Qué hace cada capa?**

| Capa | Función | Analogía |
|------|---------|----------|
| `Conv2D` | Detecta patrones con filtros | Los ojos que escanean la imagen |
| `MaxPooling2D` | Reduce el tamaño y guarda lo importante | Resumir un libro en notas clave |
| `Dropout(0.5)` | Apaga el 50% de neuronas al azar durante entrenamiento | Estudiar sin poder ver tus notas |
| `Flatten` | Convierte la matriz 2D en una lista 1D | Doblar un periódico en tira |
| `Dense(250)` | Neuronas clásicas que procesan todo | El cerebro que toma la decisión final |
| `Dense(1, sigmoid)` | Salida: un número entre 0 y 1 | 0 = Gato, 1 = Perro |

> 💡 **Dropout** es un truco genial: al "apagar" neuronas aleatoriamente durante el entrenamiento, la red no puede memorizar respuestas de memoria. Se ve obligada a **generalizar**, igual que cuando estudias sin poder ver tus apuntes.

````python
# 🏗️🤖 Construimos una Red Neuronal Convolucional (CNN) con tres capas convolucionales para extraer características de las imágenes,
# capas de MaxPooling para reducir dimensiones y una capa densa final con activación sigmoide para clasificar entre perro 🐶 y gato 🐱.
# También agregamos Dropout para evitar sobreajuste y asegurarnos de que el modelo generalice bien. 🚀🔥
modeloCNN = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(tamano_img, tamano_img, 1)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(250, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

modeloCNN.summary()
print('\n✅ ¡Arquitectura de la red neuronal creada exitosamente!')
````

# (Paso 6) 🏋️ ¡Entrenando a nuestra IA: De aprendiz a experto! 🔥

Ha llegado el momento clave. Vamos a **enseñarle** a nuestra red neuronal.

## ¿Cómo aprende una red neuronal?

El proceso de aprendizaje funciona así:

```
1. La red ve una imagen → hace una predicción (ej: "Perro")
2. Comparamos con la respuesta correcta (ej: "Gato" → ¡Error!)
3. El optimizador Adam calcula CUÁNTO se equivocó (eso es el "loss")
4. Ajusta millones de parámetros internos para cometer menos error
5. Repite con la siguiente imagen... ¡23,000 veces por época!
```

## ¿Qué significa cada parámetro?

| Parámetro | Valor | Significado |
|-----------|-------|-------------|
| `optimizer='adam'` | Adam | Algoritmo que decide cuánto ajustar cada parámetro |
| `loss='binary_crossentropy'` | — | Mide el error entre predicción y realidad (0=perfecto) |
| `metrics=['accuracy']` | — | El % de imágenes que clasifica correctamente |
| `batch_size=32` | 32 | Procesa 32 imágenes a la vez antes de ajustar |
| `validation_split=0.15` | 15% | Reserva el 15% de datos para verificar que no memorice |
| `epochs=2` | 2 | Cuántas veces verá TODO el dataset |

> 🎯 **¿Qué es una época?** Imagina que el dataset es un libro de texto. Una época = leer el libro completo una vez. Con 2 épocas, la IA lee el libro 2 veces. Con más épocas aprende más... pero también puede "memorizar" en lugar de entender.

````python
# 🚀🔥 ¡Preparando y entrenando a nuestra IA!
# Primero compilamos el modelo usando el optimizador 'adam' para ajustar los pesos de forma automática,
# la función de pérdida 'binary_crossentropy' ideal para clasificar entre dos clases (perro o gato),
# y medimos el desempeño con la métrica 'accuracy' para saber qué tan bien está aprendiendo.
modeloCNN.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# 🏋️‍♂️💻 ¡Manos a la obra! Entrenamos el modelo con nuestros datos:
# Utilizamos un tamaño de lote de 32 y separamos un 15% de los datos para validación.
# El entrenamiento se realiza por 2 épocas, lo que permitirá que nuestra IA comience a aprender a clasificar.
history = modeloCNN.fit(X, Y, batch_size=32, validation_split=0.15, epochs=2)
````

# (Paso 7) 📊 Evaluación: ¿Qué tan buena quedó nuestra IA?

Llegó el momento de calificar a nuestra IA. Como en un examen, le presentaremos imágenes que **ya conoce** y mediremos cuántas acierta.

**¿Qué significan los números?**

| Resultado | Loss | Accuracy | Interpretación |
|-----------|------|----------|----------------|
| 😬 Malo | ~1.2 | ~55% | Casi adivinando al azar (cara o cruz) |
| 🛠️ Mejorable | ~0.8 | ~70% | Aprende pero se equivoca seguido |
| ✅ Bueno | ~0.4 | ~82% | Lo que esperamos con 2 épocas |
| 🔥 Ideal | ~0.2 | ~95% | Necesitaría más épocas y arquitectura avanzada |

> 💡 **¿Por qué no entrenamos más épocas?** Con 2 épocas tenemos un buen balance entre velocidad y resultado para este taller. En proyectos reales se usan decenas o centenas de épocas con hardware especializado.

````python
# 📊🔍 ¡Momento de la verdad! Evaluamos qué tan bien aprendió nuestra IA 🧠✨
# 🎭 **Escenario 1 (MALO 😬)** → Si el loss es ~1.2 y la accuracy es ~55%, nuestra red aún está confundida, como si intentara adivinar en lugar de reconocer bien las imágenes. Necesita más entrenamiento o ajustar sus parámetros. 🔄📉
# 🤔 **Escenario 2 (MEJORABLE 🛠️)** → Si el loss es ~0.8 y la accuracy ronda el 70%, el modelo ha mejorado, pero aún se equivoca bastante. Puede estar memorizando ejemplos en lugar de aprender realmente a generalizar. 📈🔍
# 🚀 **Escenario 3 (IDEAL 🔥)** → Si el loss es ~0.2 y la accuracy está cerca del 95%, ¡bingo! 🎯 La IA ha aprendido a diferenciar perros 🐶 y gatos 🐱 con gran precisión, haciendo predicciones confiables. ¡Así sí vale la pena usarla en el mundo real! 💡🏆
loss, accuracy = modeloCNN.evaluate(X, Y)

print(f'\n📊 Resultados del modelo:')
print(f'   Loss (error):     {loss:.4f}')
print(f'   Accuracy:         {accuracy:.2%}')

if accuracy >= 0.90:
    print('\n🔥 ¡EXCELENTE! Tu modelo supera el 90% de precisión.')
elif accuracy >= 0.80:
    print('\n✅ ¡MUY BIEN! Tu modelo supera el 80% de precisión.')
elif accuracy >= 0.70:
    print('\n🛠️ MEJORABLE. Con más épocas mejoraría bastante.')
else:
    print('\n😬 Necesita más entrenamiento. ¡Prueba aumentar las épocas!')
````

> 🔥 **Este es uno de los problemas más importantes en IA real:** los modelos no saben lo que no saben. Si les llega algo fuera de su entrenamiento, igual dan una respuesta... ¡con total confianza!

Ejecuta la siguiente celda para ver qué dice nuestra IA sobre el pulpo 👇

````python
# 🐙💥 EL GRAN EXPERIMENTO: ¿Qué dirá la IA sobre el pulpo?

ruta_pulpo = '/content/PerroGatoTest/imagen9.png'

if not os.path.exists(ruta_pulpo):
    print('⚠️ La imagen del pulpo no se encontró.')
    print('   Asegúrate de que imagen9.png esté en /content/PerroGatoTest/')
else:
    # Cargamos y procesamos el pulpo IGUAL que si fuera un perro o gato
    img_pulpo = cv2.imread(ruta_pulpo)
    img_color = cv2.cvtColor(img_pulpo, cv2.COLOR_BGR2RGB)  # Para mostrar en color

    img_proc = cv2.resize(img_pulpo, (tamano_img, tamano_img))
    img_proc = cv2.cvtColor(img_proc, cv2.COLOR_BGR2GRAY)
    img_proc = img_proc.reshape(1, tamano_img, tamano_img, 1).astype('float32') / 255

    pred_pulpo = modeloCNN.predict(img_proc, verbose=0)
    valor = pred_pulpo[0][0]
    clase_pulpo = '🐶 Perro' if valor > 0.5 else '🐱 Gato'
    confianza_pulpo = valor if valor > 0.5 else 1 - valor

    # Visualización dramática
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    fig.patch.set_facecolor('#1a1a2e')

    axes[0].imshow(img_color)
    axes[0].set_title('Lo que NOSOTROS vemos:\n🐙 ¡Un pulpo!', color='white', fontsize=13, fontweight='bold')
    axes[0].axis('off')

    axes[1].imshow(img_proc[0].reshape(tamano_img, tamano_img), cmap='gray')
    axes[1].set_title(f'Lo que la IA decide:\n{clase_pulpo} ({confianza_pulpo:.0%} segura 😱)',
                      color='yellow', fontsize=13, fontweight='bold')
    axes[1].axis('off')

    plt.suptitle('🚨 ¡ALERTA! Dataset contaminado', fontsize=16, color='red', fontweight='bold')
    plt.tight_layout()
    plt.show()

    print(f'\n🐙 Imagen: PULPO')
    print(f'🤖 La IA dice: {clase_pulpo}')
    print(f'📊 Valor de salida (sigmoid): {valor:.4f}')
    print(f'🎲 Confianza reportada: {confianza_pulpo:.1%}')
````

## 💥 ¿Qué acaba de pasar? La lección más importante del día

### Nuestra IA clasificó un pulpo como perro o gato... ¡con total confianza!

Esto no es un error de programación. Es una **limitación fundamental** de cómo funcionan estos modelos.

**¿Por qué ocurre esto?**

La red neuronal **solo puede dar respuestas dentro de lo que conoce**. Fue entrenada con perros (1) y gatos (0). Su función sigmoid siempre produce un número entre 0 y 1. No tiene vocabulario para decir *"esto es un pulpo"* o *"no sé qué es esto"*.

### Las 3 lecciones de oro en IA

```
🗑️  GARBAGE IN → GARBAGE OUT
    Si metes datos basura, obtienes predicciones basura.
    Una IA es tan buena como su dataset.

📦  FUERA DE DISTRIBUCIÓN = CAOS
    Si le muestras algo que nunca vio en el entrenamiento,
    la IA responderá con confianza... aunque esté completamente equivocada.

🔍  LA IA NO SABE LO QUE NO SABE
    No tiene duda. No tiene vergüenza. No tiene sentido común.
    Por eso los humanos siempre debemos supervisar los sistemas de IA.
```

### ¿Dónde falla esto en el mundo real?

- 🚗 Un auto autónomo entrenado en calles de EE.UU. llega a México y no reconoce señales de tráfico diferentes
- 🏥 Un sistema de diagnóstico médico entrenado con pacientes adultos falla con niños
- 🏭 Un detector de defectos entrenado en verano falla en invierno por diferente iluminación

**La calidad del dataset no es un detalle técnico. Es la base de todo.**
